#### Cálculo da variância

```python
input_data = data_t.values  
print(input_data)
```

O input_data terá a matriz dos valores NumPy do data_t. Valores NumPy são valores típicos para análise estatística.

```python
variances = input_data.var(axis = 0) 
print(variances.shape)
media_v = variances.mean()
media_v
```

Com o resultado obtido, podemos verificar que o cálculo da variância ocorreu em todas as colunas do dataset, como previsto.

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
valor = VarianceThreshold(threshold = media_v * 4)
filtro = valor.fit_transform(input_data)
filtro.shape
```

Com a primeira linha vamos definir a condição de que os valores têm que ser superiores a 4 vezes a média das variâncias. Esta condição irá ser guardada na variável valor.

O filtro é responsável por aplicar essa condição à matriz input_data. Ficamos assim com apenas 296 genes, uma vez que estes são os mais significativos, ou seja, os mais interessantes para o estudo.

### ANOVA

```python
from sklearn.feature_selection import SelectPercentile, f_classif
import numpy as np
```

```python
selector = SelectPercentile(f_classif, percentile = 10)

output_ANOVA = unified_meta.values[:, 48]

selector.fit_transform(data_t, output_ANOVA)
genes_sel = np.where(selector.pvalues_ < 0.01)
print(data_t.columns[genes_sel])
```

Utilizamos um teste ANOVA para selecionar características importantes (genes) com base na relação com as classes de cancro (SUBTYPES) presentes na coluna 48. A escolha de 10% e p-valor < 0,01 irá determinar quais as características que serão mantidas após o teste. Essas características serão impressas.

O teste ANOVA identificou um conjunto de 9562 genes que apresentam uma relação estatisticamente significativa com os subtipos de cancro. Ao passarmos de inicialmente 17494 para 9562, diminuimos a dimensionalidade do conjunto de dados, permanecendo os genes mais relevantes para o estudo dos subtipos de cancro.

### PCA

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 10)
X_r = pca.fit(data_sc).transform(data_sc)

print('Explained Variance: %s'% str(pca.explained_variance_ratio_)) #Imprime a proporção da variância explicada por cada componente principal.

pca.explained_variance_ratio_.sum()
```

```python
n = 10
plt.bar(range(n), pca.explained_variance_ratio_ * 100)
plt.xticks(range(n), ['PC' + str(i) for i in range(1, n + 1)])
plt.title("Variância explicada por PC")
plt.ylabel("Percentagem")
plt.show()
```

O PCA é uma técnica estatística utilizada para simplificar a complexidade num conjunto de dados, reduzindo o número de variáveis enquanto mantém as informações essenciais contidas nos dados. 

Pela análise do gráfico, verificamos que o PC1 é o que apresenta maior variância, tal como esperado.

### Clustering Hierárquico

```python
from sklearn.cluster import AgglomerativeClustering

hclust = AgglomerativeClustering(n_clusters = 4, metric = "euclidean", linkage = "ward") #ward: minimiza a variância entre os clusters - escolha por defeito
hclust.fit(data_sc)                                       #mede similaridades consoante a distância gráfica

output_cluster = unified_meta.values[:, 48] 
res_hc = hclust.labels_
pd.crosstab(output_cluster, res_hc)
```

A tabela resultante representa o número de amostras, dos diferentes subtipos, armazenadas nos 4 clusters. 

Podemos concluir que no cluster 0, os subtipos UCEC_CN_LOW e UCEC_MSI são aqueles que mais se destacam, ou seja aqueles que apresentam uma maior similaridade.

### K-Means

```python
from sklearn.cluster import KMeans

kmeans_est = KMeans(init='random', n_clusters = 4, n_init = 100)
res_kmeans = kmeans_est.fit(data_sc).predict(data_sc)

pd.crosstab(output_ANOVA, res_kmeans)
```

O K-Means é um algoritmo que visa particionar um conjunto de dados em K clusters, onde cada ponto de dados pertence ao cluster cujo ponto médio é mais próximo. 

Desta forma, para o subtipo UCEC_CN_HIGH, o cluster 0 é o que apresenta maior presença significativa e para os subtipos UCEC_CN_LOW, UCEC_MSI, UCEC_POLE o cluster 1 é o que apresenta maior presença significativa.

Irá ser feito o treino de 2/3 dos dados e usado 1/3 para teste. 

```python
num_tst = int(data_t.shape[0] / 3) #divide os dados em 3 
indices = np.random.permutation(data_t.shape[0])


train_in = data_sc[indices[:-num_tst]]
train_out = outputs[indices[:-num_tst]]
test_in  = data_sc[indices[-num_tst:]]
test_out = outputs[indices[-num_tst:]]
```
