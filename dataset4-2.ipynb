### Variáveis escolhidas

Para a análise do estudo, selecionou-se como variáveis o Subtype, Grade, Age e DFS_Status (Disease Free Status). Antes de procedermos à criação dos gráficos, temos de fazer um processamento.

#### Processamento da Subtype

```python
unified_meta['SUBTYPE'].isnull().sum()
```

#### Processamento da Grade

```python
unified_meta['GRADE'].isnull().sum()
```

Logo, não existem valores nulos.

#### Processamento da Age

```python
unified_meta['AGE'].isnull().sum()
```

```python
unified_meta = unified_meta.dropna(subset=['AGE'])
unified_meta['AGE'].isnull().sum()
```

#### Processamento da DFS_STATUS

```python
unified_meta['DFS_STATUS'].isnull().sum()
```

```python
unified_meta = unified_meta.dropna(subset=['DFS_STATUS'])
unified_meta['DFS_STATUS'].isnull().sum()
```

Agora que as variáveis escolhidas estão tratadas, procedeu-se à criação dos gráficos.

#### Gráficos

```python
unified_meta.groupby("SUBTYPE").size().plot.pie(title = 'Subtype', ylabel = '')
```

Pela análise do gráfico, verifica-se que o subtipo de cancro mais comum é o 'UCEC_CN_LOW' e menos é o 'UCEC_POLE'.

```python
unified_meta.groupby("GRADE").size().plot.pie(title = 'Grade', ylabel = '')
```

O grau principal dos resultados é o grau 3.

```python
unified_meta.groupby("AGE").size().plot.pie(title = 'Age', ylabel = '')
```

Devido ao gráfico ser pouco intuitivo, procedeu-se à criação de um gráfico de barras com intervalos das idades: [0,25], [25,40], [40,65], [65, 100]. 

```python
unified_meta['AGE'].dtypes
```

Verificou-se que a variável AGE é do tipo Object, contudo precisamos que esta seja do tipo inteiro para procedermos à criação do gráfico de barras

```python
unified_meta = unified_meta.astype({'AGE':'int'})
```

```python
unified_meta.groupby([pd.cut(unified_meta.AGE,[0,25,40,65,100])]).size().plot.bar() 
```

De maneira a fazer uma melhor distribuição das idades, reajustou-se os intervalos: [30,40], [40,50], [50,65], [65,80], [80,100].

```python
unified_meta.groupby([pd.cut(unified_meta.AGE,[30,40,50,65,80,100])]).size().plot.bar() 
```

A faixa etária mais comum dos pacientes é dos 50 aos 65 anos.

```python
unified_meta.groupby("DFS_STATUS").size().plot.pie(title = 'Disease Free Status', ylabel = '')
```

Indica que a maior parte dos pacientes não apresentou reocorrência da doença.

```python
unified_meta.groupby(['GRADE','DFS_STATUS']).size().plot.bar()
```

É possível concluir que o grau 3 é o que apresenta maior reocorrência depois do tratamento, mas também é o que apresenta maior sucesso nos tratamentos. Tal é comprovado com o gráfico obtido anteriormente, que mostra que o grau 3 é o grau mais comum dos pacientes em estudo.

```python
unified_meta.groupby(['DFS_STATUS',pd.cut(unified_meta.AGE,[30,40,50,65,80,100])]).size().plot.bar()
```

A faixa etária com mais reocorrência é a dos 50 aos 65. Contudo, é aquela que apresenta maior sucesso nos tratamentos.

Já a faixa etária dos 30 aos 40, não apresenta reincidência após tratamentos.

```python
unified_meta.groupby(['SUBTYPE','GRADE']).size().plot.bar()
```

O grau 3 é aquele que apresenta maior incidência nos subtipos UCEC_CN_HIGH e UCEC_POLE. Contudo, no subtype UCEC_CN_LOW é o menor dos 3 mais significativos.

```python
unified_meta.groupby(['SUBTYPE',pd.cut(unified_meta.AGE,[30,40,50,65,80,100])]).size().plot.bar()
```

Para o subtype UCEC_UC_HIGH, a faixa etária com maior incidência é a dos 65 aos 80. 

No subtype UCEC_UC_LOW, a faixa etária com maior incidência é a dos 50 aos 65. 

No subtype UCEC_MSI, a faixa etária com maior incidência é a dos 50 aos 65. 

Por último, na subtype UCEC_POLE, a faixa etária com maior incidência é a dos 50 aos 65. 

Podemos concluir que, na faixa etária dos 50 aos 65 anos se verifica uma grande incidência de vários subtipos de cancro, pois como vimos anteriormente é a faixa etária que apresenta maior número de pacientes.
